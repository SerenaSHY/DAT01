### Types of Data Visualization in Matplotlib

### 1. Line Plot

The line plot, a cornerstone of data visualization, excels at revealing trends over time or across different categories. It connects data points with a line, visually depicting how a value changes in relation to another variable.

```python
# Data for number of orders per month
data = {
    "Months": ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"],
    "No. Of Orders for Jacket": [8, 10, 12, 14, 18, 20, 26, 24, 22, 14, 12, 9],
    "No. Of Orders for Summer Clothes": [25, 27, 24, 20, 15, 13, 9, 13, 14, 16, 18, 22]
}
```

```python
# new code
order_per_month = pd.DataFrame(data)
order_per_month
```

```python
# Size of the graph [width, height]
plt.figure(figsize=(8, 6))

# [y, x] axis
# marker o = oval
#  markers = square
# linestyle: solid(default), dashed, dotted, dashdot
plt.plot(order_per_month["Months"], order_per_month["No. Of Orders for Jacket"], 
         marker = 'o', color = 'pink', linestyle = 'dashed', linewidth = 2, label = "No. Of Orders for Jacket")
plt.plot(order_per_month["Months"], order_per_month["No. Of Orders for Summer Clothes"], 
         marker = "s", color = "red", linestyle = 'dotted', linewidth = 2, label = "No. Of Orders for Summer Clothes")

# alpha: opacity
plt.fill_between(order_per_month["Months"], order_per_month["No. Of Orders for Summer Clothes"], 
                 color = 'grey', alpha = 0.4, label = "No. Of Orders for Summer Clothes")
plt.fill_between(order_per_month["Months"], order_per_month["No. Of Orders for Jacket"], 
                 color = 'skyblue', alpha = 0.4, label = "No. Of Orders for Jacket")

# Adding a label
plt.xlabel("Months", fontsize = 12, fontweight = 'bold')
plt.ylabel("No. Of Orders", fontsize = 12, fontweight = 'bold')
# Add a title
plt.title("No. Of Orders per Month", fontsize = 14, fontweight = 'bold')
# Add the legend
# loc = locatio of the legend, upper/lower/center (upper right is the default w/o the [loc=])
# put the legend outside by using "bbox_to_anchor" parameter for .legend
plt.legend(bbox_to_anchor =(0.5,-0.40), loc='lower center')
plt.tight_layout()
# tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area.

# Display the lineplot
plt.show()
```

### 2. Pie Charts

Pie charts, those delightful circles sliced into wedges, are a go-to choice for visualizing proportions of a whole.  Matplotlib's pie chart functionality, accessible through the pyplot.pie() function, empowers you to represent data categories and their relative sizes in a clear and intuitive way.

```python
# Data for spending amounts for each category
data = {
    "Categories": ['Rent', 'Food', 'Transportation', 'Savings', 'Other'],
    "Amounts": [500, 300, 200, 100, 100]
}
```

```python
# New code
spending_amount_df = pd.DataFrame(data)
spending_amount_df
```

```python
explodes = [0.1, 0, 0, 0, 0]
#Emphasize the biggest slice and separated from the rest, explode=explodes is required

plt.pie(spending_amount_df["Amounts"], labels=spending_amount_df["Categories"], 
        autopct='%.2f%%', colors=["#947EB0", "violet", "#FF0000", (0,0,1), (0,1,1)], explode=explodes) 
# 5 colours for each slices

# or without explodes = [0.1, 0, 0, 0, 0] at line 1
# plt.pie(spending_amount_df["Amounts"], labels=spending_amount_df["Categories"], 
        # autopct='%.2f%%', colors=["#947EB0", "violet", "#FF0000", (0,0,1), (0,1,1)], explode=[0.1, 0, 0, 0, 0]) 

# Colors: use coolors.co to search the colour charts
# 1. Named Colors 
# 2. Hexadecimal 
# 3. RGB Tuples

plt.title("Budget per each Category", fontsize=14, fontweight='bold')

# bbox_to_anchor = (x, y)
plt.legend(labels=spending_amount_df["Categories"], 
           title="Categories", loc="lower right", bbox_to_anchor =(1.5,0.5))

# plt.tight_layout()
plt.show()
```

#### 3. Bar Charts

Bar charts, the workhorses of data visualization, excel at showcasing comparisons between different categories.  Matplotlib's bar chart functionality, primarily accessed through the pyplot.bar() function, allows you to translate numerical data into a series of rectangular bars. The height of each bar corresponds to the value it represents, while the bars are positioned along a horizontal axis that typically represents the categories being compared.

```python
# Data for student enrollment per course
# \n - create a new line
data = {
    "Course": ['Data \nAnalytics', 'Frontend \nDevelopment', 'DevOps', 'Full-Stack Development'],
    "Students": [30, 25, 32, 28]
}
students_per_course_df = pd.DataFrame(data)
students_per_course_df
```

```python
# New code
students_per_course_df = pd.DataFrame(data)
students_per_course_df
```

```python
plt.bar(students_per_course_df["Course"], students_per_course_df["Students"], 
        color=["#b5e2fa", "#eddea4", "#f7a072", "#0fa3b1"], width=0.8) # Width: width of bar
plt.xlabel("Courses", fontsize=12, fontweight="bold")
plt.ylabel("Students", fontsize=14, fontweight="bold")

plt.title("Students enrolled per Course", fontsize=14, fontweight="bold")

plt.xticks(rotation=45, fontsize = 10)
plt.show()
```
